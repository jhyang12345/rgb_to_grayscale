```python
from PIL import Image
import os
import csv
```

```python
maindirectory = os.getcwd()
print(os.getcwd())
#os.chdir("./python_images/iu")#set current directory to the directory of images
```

```python
os.chdir(maindirectory + "/python_images/iu")
imagelist = sorted(os.listdir())
rgblist = []

for image in imagelist[:4]:
    try:
        im = Image.open(image)
        pix = im.load()        
        x = im.size[0]#gettin x
        y = im.size[1]#getting y
        print(im)
        for i in range(x):
            for j in range(y):
                rgblist.append(pix[i, j])
    except:
        continue
        
```

```python
with open("output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rgblist)
```

```python
if not os.path.exists("../gray_scale"):#make directory to put gray_scale images
    os.makedirs("../gray_scale")
```

```python
imagelist = sorted(os.listdir())
header = "../gray_scale/"
for image in imagelist:
    try:
        img = Image.open(image).convert("L")        
        file_name, ext = image.split(".")
        img.save(header + file_name + "_gray" + "." + ext)
        print(img)
    except Exception as inst:
        print(inst)
        continue
```

```python
gray_images = os.listdir("../gray_scale")
header = "../gray_scale/"
graylist = []
for image in imagelist[:4]:
    imagename, ext = image.split(".")
    
    try:
        im = Image.open(header + imagename + "_gray." + ext)
        print(im)
        pix = im.load()        
        x = im.size[0]#gettin x
        y = im.size[1]#getting y
        
        for i in range(x):
            for j in range(y):
                graylist.append([pix[i, j]])#grayscale images output only the grayscale
    except Exception as inst:
        print(inst)
        continue
        
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
#import necessary modules
```

```python
#attempting to handle csv files with tensorflow
os.chdir(maindirectory)
inputfilename_queue = tf.train.string_input_producer(["./python_images/gray_scale/output.csv"])
outputfilename_queue = tf.train.string_input_producer(["./python_images/iu/output.csv"])
reader = tf.TextLineReader()
key, value = reader.read(inputfilename_queue)
key, output = reader.read(outputfilename_queue)

record_defaults = [[1], [1], [1]]
col1, col2, col3 = tf.decode_csv(value, record_defaults=record_defaults)
record_defaults = [[1]]
col4 = tf.decode_csv(output, record_defaults=record_defaults)

features = tf.stack([col1, col2, col3])

output = tf.stack([col4])

with tf.Session() as sess:
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(coord=coord)
    
    for i in range(1200):
        example, label = sess.run(features, output)
    coord.request_stop()
    coord.join(threads)
```
