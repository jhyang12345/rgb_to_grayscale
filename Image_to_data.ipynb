```python
from PIL import Image
import os
import csv
```

```python
print(os.getcwd())
os.chdir("./python_images/iu")#set current directory to the directory of images
```

```python
imagelist = os.listdir()
rgblist = []

for image in imagelist[:4]:
    try:
        im = Image.open(image)
        pix = im.load()        
        x = im.size[0]#gettin x
        y = im.size[1]#getting y
        for i in range(x):
            for j in range(y):
                rgblist.append(pix[i, j])
    except:
        continue
        
```

```python
with open("output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rgblist)
```

```python
if not os.path.exists("../gray_scale"):#make directory to put gray_scale images
    os.makedirs("../gray_scale")
```

```python
imagelist = os.listdir()
header = "../gray_scale/"
for image in imagelist:
    try:
        img = Image.open(image).convert("L")        
        file_name, ext = image.split(".")
        img.save(header + file_name + "_gray" + "." + ext)
        print(image)
    except Exception as inst:
        print(inst)
        continue
```

```python
gray_images = os.listdir("../gray_scale")
header = "../gray_scale/"
graylist = []

for image in gray_images[:4]:
    try:
        im = Image.open(header + image)
        print(im)
        pix = im.load()        
        x = im.size[0]#gettin x
        y = im.size[1]#getting y
        
        for i in range(x):
            for j in range(y):
                graylist.append([pix[i, j]])#grayscale images output only the grayscale
    except Exception as inst:
        print(inst)
        continue
        
```

```python
print(graylist[:100])
with open("../gray_scale/output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(graylist)
```
