```python
import tensorflow as tf
import pandas as pd
import numpy as np
#import necessary modules
```

```python
print(os.getcwd())

os.chdir(maindirectory)#Changing back to python_images directory
output =pd.read_csv("./python_images/gray_scale/output.csv")
features = pd.read_csv("./python_images/iu/output.csv")
```

```python
print(features.head(10))
print(output.head(10))
print(features.index)
print(output.index)

#print dataframe head and translate dataframe to matrix
features = features.as_matrix()
output = output.as_matrix()
```

```python
features = features.astype(np.float32)
output = output.astype(np.float32)

print(features.shape)
print(output.shape)
print(features[:10])
print(output[:10])
```

```python
x_ = tf.placeholder(tf.float32, shape=[None, 3], name="x-input")#4457199, 3
y_ = tf.placeholder(tf.float32,shape=[None], name="y-input")
```

```python

Theta1 = tf.Variable(tf.random_uniform([3, 16], -2, 2), name="Theta1")
Theta2 = tf.Variable(tf.random_uniform([16, 10], -2, 2), name="Theta2")
Theta3 = tf.Variable(tf.random_uniform([10, 1], -2, 2), name="Theta3")
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(Theta1))
```

```python
Bias1 = tf.Variable(tf.zeros([16]), name="Bias1")
Bias2 = tf.Variable(tf.zeros([10]), name="Bias2")
Bias3 = tf.Variable(tf.zeros(1), name="Bias3")
```

```python
A2 = tf.nn.sigmoid(tf.matmul(x_, Theta1) + Bias1)
A3 = tf.nn.relu(tf.matmul(A2, Theta2) + Bias2)
Output = tf.matmul(A3, Theta3) + Bias3
```

```python
cost = tf.reduce_mean(tf.square(y_ - Output))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
for j in range(20):
    feature_batch = []
    output_batch = []
    
    #print(sess.run(Theta1))
    for i in np.random.choice(len(features), 500):
        feature_batch.append(features[i])
        output_batch.append(output[i][0])        
    sess.run([cost, Output], feed_dict={x_: feature_batch, y_ : output_batch})
    feature_batch = np.array(feature_batch)
    output_batch = np.array(output_batch)
    sess.run(train_step, feed_dict={x_: feature_batch, y_ : output_batch})
    print(sess.run(cost, feed_dict={x_: feature_batch, y_: output_batch}))
    #print(sess.run(Output, feed_dict={x_: feature_batch, y_: output_batch}))
    #print(sess.run(y_, feed_dict={x_: feature_batch, y_: output_batch[:,None]}))
    
#for i in range(len(features)):
    #feature_batch, label_batch = tf.train.shuffle_batch([features, output], batch_size=40, capacity=5000, min_after_dequeue=1000)
    #for i in range(10000):
#    sess.run(train_step, feed_dict={x_: features[i][None,:], y_ : y[i]})
       #print(sess.run(Output, feed_dict={x_: features[:1000]}))
       #print(sess.run(cost, feed_dict={x_: features[0][None,:], y_ : output[0][None,:]}))    


    #print(sess.run(Output, feed_dict={x_: features[:10000]}))
```
