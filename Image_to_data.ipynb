```python
import tensorflow as tf
import pandas as pd
import numpy as np
#import necessary modules
```

```python
print(os.getcwd())

os.chdir(maindirectory)#Changing back to python_images directory
output =pd.read_csv("./python_images/gray_scale/output.csv")
features = pd.read_csv("./python_images/iu/output.csv")
```

```python
print(features.head(10))
print(output.head(10))
print(features.index)
print(output.index)

#print dataframe head and translate dataframe to matrix
features = features.as_matrix()
output = output.as_matrix()
```

```python
features = features.astype(np.float32)
output = output.astype(np.float32)

print(features.shape)
print(output.shape)
print(features[:10])
print(output[:10])
```

```python
epochs = 1000
learning_rate = 0.001
display_step = 5
```

```python
x_ = tf.placeholder(tf.float32, shape=[None, 3], name="x-input")#4457199, 3
y_ = tf.placeholder(tf.float32, shape=[None, 1], name="y-input")
```

```python
Theta1 = tf.Variable(tf.random_uniform([3, 16], -2, 2), name="Theta1")
Theta2 = tf.Variable(tf.random_uniform([16, 10], -2, 2), name="Theta2")
Theta3 = tf.Variable(tf.random_uniform([10, 1], -2, 2), name="Theta3")
```

```python
Bias1 = tf.Variable(tf.zeros([1, 16]), name="Bias1")
Bias2 = tf.Variable(tf.zeros([1, 10]), name="Bias2")
Bias3 = tf.Variable(tf.zeros([1, 1]), name="Bias3")
```

```python
layer_one = tf.add(tf.matmul(x_, Theta1), Bias1)
layer_two = tf.add(tf.matmul(layer_one, Theta2), Bias2)
pred = tf.add(tf.matmul(layer_two, Theta3), Bias3)
cost = tf.reduce_sum(tf.pow(pred - y_, 2)) / 500
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
for j in range(epochs):
    feature_batch = []
    output_batch = []
    
    #print(sess.run(Theta1))
    for i in np.random.choice(len(features), 500):
        feature_batch.append(features[i])
        output_batch.append(output[i])
    feature_batch = np.asarray(feature_batch)
    output_batch = np.asarray(output_batch)
    sess.run(optimizer, feed_dict={x_: feature_batch, y_: output_batch})
    run_cost = sess.run(cost, feed_dict={x_: feature_batch, y_: output_batch})
    print(run_cost)
    
```

```python
epochs = 500
learning_rate = 0.001
display_step = 5
```

```python
x_ = tf.placeholder(tf.float32, shape=[None, 3], name="x-input")#4457199, 3
y_ = tf.placeholder(tf.float32, shape=[None, 1], name="y-input")
```

```python
Theta1 = tf.Variable(tf.random_uniform([3, 8], -1, 1), name="Theta1")
Theta2 = tf.Variable(tf.random_uniform([8, 8], -1, 1), name="Theta2")
Theta3 = tf.Variable(tf.random_uniform([8, 1], -1, 1), name="Theta3")
```

```python
Bias1 = tf.Variable(tf.random_uniform([1, 8], -1, 1), name="Bias1")
Bias2 = tf.Variable(tf.random_uniform([1, 8], -1, 1), name="Bias2")
Bias3 = tf.Variable(tf.random_uniform([1, 1], -1, 1), name="Bias3")
```

```python
layer_one = tf.add(tf.matmul(x_, Theta1), Bias1)
layer_two = tf.add(tf.matmul(layer_one, Theta2), Bias2)
pred = tf.add(tf.matmul(layer_two, Theta3), Bias3)
cost = tf.reduce_sum(tf.pow(pred - y_, 2))
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(epochs):
        for (x, y) in zip(features[:1000], output[:1000]):
            input_x = np.asarray([x])
            input_y = np.asarray([y])
            sess.run(optimizer, feed_dict={x_: input_x, y_: input_y})
        p_cost = sess.run(cost, feed_dict={x_: features[:10], y_: output[:10]})
        print(p_cost)
        t_pred = sess.run(pred, feed_dict={x_ :features[:10], y_: output[:10]})
        print(t_pred)
        print(output[:10])
    saver.save(sess, "checkpoints/image_params")
```
