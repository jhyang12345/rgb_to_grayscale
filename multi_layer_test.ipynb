```python
import tensorflow as tf
import numpy as np
```

```python
def f(a, b, c):
    return 3 * a + 2 * b - c
input = []
output = []
for x in range(1000):
    a = np.random.random_sample() * 100
    b = np.random.random_sample() * 100
    c = np.random.random_sample() * 100
    input.append([a, b, c])
    output.append([f(a, b, c)])
```

```python
saver = tf.train.Saver()
learning_rate = 0.001
epochs = 1000
display_step = 5
```

```python
train_X = np.asarray(input)
train_Y = np.asarray(output)
test_X = np.asarray(input[-300:])
test_Y = np.asarray(output[-300:])
n_samples = train_X.shape[0]
print(n_samples)
print(train_X.shape)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3], name="x-input")
Y = tf.placeholder(tf.float32, shape=[None, 1], name="y-input")
```

```python
W1 = tf.Variable(tf.random_uniform([3, 3], -1, 1), name="weight1")
b1 = tf.Variable(tf.random_uniform([1, 3], -1, 1), name="bias1")
W2 = tf.Variable(tf.random_uniform([3, 1], -1, 1), name="weight2")
b2 = tf.Variable(tf.random_uniform([1, 1], -1, 1), name="bias2")
```

```python
layer_one = tf.add(tf.matmul(X, W1), b1)
pred = tf.add(tf.matmul(layer_one, W2), b2)
cost = tf.reduce_sum(tf.pow(pred - Y, 2)) / 2 * n_samples
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
init = tf.global_variables_initializer()


with tf.Session() as sess:
    sess.run(init)
    for epoch in range(epochs):
        for (x, y) in zip(train_X, train_Y):
            input_X = np.asarray([x])
            input_Y = np.asarray([y])
            sess.run(optimizer, feed_dict={X: input_X, Y: input_Y})

    for (x, y) in zip(test_X, test_Y):
        input_X = np.asarray([x])
        input_Y = np.asarray([y])
        test_output = sess.run(pred, feed_dict={X: input_X, Y: input_Y})
        print("Results:", input_Y, test_output)
    matrix_output = sess.run(pred, feed_dict={X: test_X[:10], Y: test_Y})
    print(test_Y[:10])
    print(matrix_output)
    saver.save(sess, "./tmp/multi_layer.ckpt")
```
