```python
import tensorflow as tf
import numpy as np
```

```python
def f(a, b, c):
    return 3 * a + 2 * b + c
input = []
output = []
for x in range(500):
    a = np.random.random_sample()
    b = np.random.random_sample()
    c = np.random.random_sample()
    input.append([a, b, c])
    output.append([f(a, b, c)])
    
```

```python
learning_rate = 0.001
epochs = 1000
display_step = 5
```

```python
train_X = np.asarray(input)
train_Y = np.asarray(output)
test_X = np.asarray(input[-300:])
test_Y = np.asarray(output[-300:])
n_samples = train_X.shape[0]
print(n_samples)
print(train_X.shape)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3], name="x-input")
Y = tf.placeholder(tf.float32, shape=[None, 1], name="y-input")
```

```python
W = tf.Variable(tf.random_uniform([3, 1], -1, 1), name="weight")
b = tf.Variable(tf.random_uniform([1, 1], -1, 1), name="bias")
```

```python
pred = tf.add(tf.matmul(X, W), b)
cost = tf.reduce_sum(tf.pow(pred - Y, 2)) / 2 * n_samples
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
saver = tf.train.Saver()

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(epochs):
        for (x, y) in zip(train_X, train_Y):
            input_X = np.asarray([x])
            input_Y = np.asarray([y])
            sess.run(optimizer, feed_dict={X: input_X, Y: input_Y})
        if (epoch+1) % display_step == 0:
            c = sess.run(cost, feed_dict={X: train_X, Y:train_Y})
            print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(c), \
                "W=", sess.run(W), "b=", sess.run(b))
    save_path = saver.save(sess, "/tmp/model.ckpt")
            
```
